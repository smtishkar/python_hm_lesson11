```python
import sympy as sp
```

```python
x = sp.Symbol('x')
expr =  sp.sin(x)**2 - sp.cos(x)**2
expr
```

```python
expr_solve = sp.solveset(expr)
expr_solve
```

```python
expr_diff = sp.diff(expr)
expr_diff 
```

```python
graph = sp.plot(expr)
```

```python
rise_exp = sp.solve(expr_diff>0)
rise_exp
```
